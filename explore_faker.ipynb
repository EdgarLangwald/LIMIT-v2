# Faker Library Demo
Quick exploration of the [Faker](https://faker.readthedocs.io/) library for generating synthetic data.

```python
from faker import Faker

fake = Faker()
```

```python
# Basic fields
for _ in range(5):
    name = fake.name()
    city = fake.city()
    job = fake.job()
    dob = fake.date_of_birth(minimum_age=20, maximum_age=80).strftime('%d %B %Y')
    print(f"{name} was born on {dob}, lives in {city}, and works as a {job}.")
```

```python
# Company & contact details
for _ in range(5):
    name = fake.name()
    company = fake.company()
    email = fake.email()
    phone = fake.phone_number()
    print(f"{name} works at {company}. Reach them at {email} or {phone}.")
```

```python
# Addresses
for _ in range(3):
    name = fake.name()
    address = fake.address().replace('\n', ', ')
    print(f"{name} lives at: {address}")
```

```python
# Multi-locale: German names and addresses
fake_de = Faker('de_DE')

for _ in range(5):
    name = fake_de.name()
    city = fake_de.city()
    job = fake_de.job()
    print(f"{name} aus {city} arbeitet als {job}.")
```

```python
# Text & color curiosities
for _ in range(3):
    color = fake.color_name()
    plate = fake.license_plate()
    sentence = fake.sentence(nb_words=8)
    print(f"[{plate}] ({color} car) — {sentence}")
```
